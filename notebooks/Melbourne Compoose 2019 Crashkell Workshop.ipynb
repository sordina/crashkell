```haskell
multiple3 x = mod x 3
multiple5 x = mod x 5

euler f 0 = 0
euler f n | multiple3 n = f n (euler f (pred n))
          | multiple5 n = f n (euler f (pred n))
          | otherwise   = euler f (pred n)
          
problem1 = euler +

problem1 10000
```

### Exercise: Sorting

```haskell
mergeSort :: [Integer] -> [Int]
mergeSort x =
    if length x < 2
      then x
      else merge (mergeSort l) (mergeSort r)
    where
    (l, r) = split x

split :: [a] -> ([a],[b])
split x =
    case length x of
      0 -> ([],[])
      1 ->
        let h = head x
            t = tail x
            s = split t
            l = fst s
            r = snd s
            in (h:l, r)
      otherwise ->
        let h1 = head x
            h2 = head (tail x)
            t  = tail (tail x)
            s  = split t
            l  = fst s
            r  = snd s
            in (h1:l, h2:r)

merge :: [Integer] -> [Integer] -> [Integer]
merge l r
  | null r = l
  | null l = r
  | otherwise =
      let lh = head l
          lt = tail l
          rh = head r
          rt = tail r
        in if lh < rh
             then lh : merge lt r
             else rh : merge l rt
                    
mergeSort [1,99,2,3,4,5,88,6]
```

<span class="hint">
    Your types must line up!
</span>

### Exercise: Indexing!

```haskell
indexA e l = let h = head l
                 t = tail l
               in if e == h
                    then 0
                    else 1 + indexA e t

indexA 7 []
```

```haskell
-- Much cooler way to find the index of an element

indexB e l = helper indexed
    where
    indexed = map (is e) (zip [0..] l)
    
is e x = e == snd x

helper l | snd (head l) = 0
         | otherwise    = 1 + helper (tail l)

indexB 'm' ['a'..'z']
```

```haskell
indexC e l = snd $ head $ mergeSort (map (distance e) (zip [0..] l))

distance e x = abs (e - snd x)

indexC 55 [21..211]
```

<span class="hint">
    The distance function should include the index.
</span>

### Exercise: Sorting with Patterns

```haskell
-- Once this is all fixed it's much nicer isn't it?

mergeSort :: [Integer] -> [Int]
mergeSort []  = []
mergeSort [x] = [x]
mergeSort x   = merge (mergeSort l) (mergeSort r)
    where
    (l, r) = split x

split :: [a] -> ([a],[b])
split (h1:h2:t) = let (l,r) = split t in (h1:l, h2:r)
split x         = (x,[])

merge :: [Integer] -> [Integer] -> [Integer]
merge l [] = l
merge [] r = r
merge (lh:lt) (rh:rt) | lh < rh   = lh : merge lt r
                      | otherwise = rh : merge l rt

mergeSort [1,99,2,3,4,5,88,6]
```

```haskell
import Control.Monad.State
import System.Random
import qualified Data.Set
import Data.List (sortBy)
import Data.Ord (comparing)

type World = Data.Set.Set (Int,Int)

randomRs :: (Int,Int) -> StdGen -> [Int]
randomRs r s = let (a, g) = randomR r s in a : randomRs r g

getClosed (x,y) w = a ++ b ++ l ++ r
    where
    l = if x > 0      && not (Data.Set.member (x-1,y) w) then [(x-1,y)] else []
    r = if x < width  && not (Data.Set.member (x+1,y) w) then [(x+1,y)] else []
    b = if y > 0      && not (Data.Set.member (x,y-1) w) then [(x,y-1)] else []
    a = if y < height && not (Data.Set.member (x,y+1) w) then [(x,y+1)] else []

fun :: [Int] -> (Int, Int) -> State World ()
fun xs c = do
    w <- get
    if length (getClosed c (Data.Set.insert c w)) < 3
    then return ()
    else do
        burn c
        w <- get
        let (xs', closed) = shuffle xs $ getClosed c w
        mapM_ (fun xs') closed
    
burn c = Data.Set.insert c
    
shuffle :: [Int] -> [a] -> ([Int], [a])
shuffle xs l = (drop 4 xs, map snd $ sortBy (comparing fst) $ zip xs l)

width = 100
height = 30

renderMaze s = unlines $ map (renderRow s) [0 .. height]
renderRow  s y = map (\x -> toChar $ Data.Set.member (x, y) s) [0 .. width]

toChar True = ' '
toChar False = 'â–ˆ'

do
    foo@(x : y : rs) <- fmap (randomRs (0, width * height)) newStdGen
    putStrLn $ renderMaze $ flip execState Data.Set.empty $ do
        fun rs (x `mod` width, y `mod` height)
        burn (0, height `div` 2)
        burn (width, height `div` 2)
```
